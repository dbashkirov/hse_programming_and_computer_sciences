#### Наибольшая возрастающая подпоследовательность

Есть последовательность чисел $a_1, a_2, a_3 ... a_n$. Нужно найти ее подпоследовательность наибольшей длины, такую что: $∀i < j: a_{k_{i}} < a_{k_{j}}$

* $dp[i]$ ─ длина НВП с концом в $i$
* $dp[0] = 1$
* $dp[i] = max_{j<i}^{a[j] < a[i]}dp[j] + 1$
* По возрастанию $i$
* $max(dp[i])$

#### Наибольшая общая подпоследовательность

Есть последовательность чисел $a_1, a_2, a_3 ... a_n$ и последовательность чисел $b_1, b_2, b_3 ... b_m.$ Нужно найти их общую подпоследовательность наибольшей длины

* $dp[i][j]$ ─ длина НОП $A[0..i-1]$ и $B[0..j-1]$
* $dp[0][0] = 0$
* Если $a[i−1] = b[j−1]$ $dp[i][j] = dp[i−1][j−1] + 1$, иначе $dp[i][j] = max(dp[i−1][j], dp[i][j−1])$
* По возрастанию $i$, По возрастанию $j$
* $dp[n][m]$

#### Задача о рюкзаке

Есть рюкзак вместимостью $W$. Также есть n предметов с весом $w_i$. Нужно максимально заполнить рюкзак

* $dp[i][j]$ ─ можно ли из первых $i$ предметов набрать суммарный вес $j$
* $dp[0][j] = False, dp[0][0] = True$
* $dp[i][j] = dp[i-1][j] \ or\  dp[i-1][j- w_i]$
* По возрастанию $i$, По возрастанию $j$
* $max_j: dp[n][j] = True$

Добавим предметам стоимость. Теперь мы хотим сделать так, чтобы суммарная стоимость взятых предметов была максимальна

* $dp[i][j]$ ─ максимальная стоимость, с которой можно из первых $i$ предметов набрать суммарный вес $j$
* $dp[0][j] = -INF, dp[0][0] = 0$
* $dp[i][j] = max(dp[i-1][j], dp[i-1][j- w_i] + с_i)$
* По возрастанию $i$, По возрастанию $j$
* $max(dp[n][j])$

#### Расстояние Левенштейна

* $dp[i][j]$ - расстояние между префиксами $t[:i]$ и $s[:j]$
* $dp[0][0] = 0, dp[i][0] = i, dp[0][j] = j$
* $dp[i][j] = min(dp[i - 1][j] + 1, dp[i][j - 1] + 1, dp[i - 1][j - 1] + chr\_diff(s[i - 1], t[j - 1]))$
* По возрастанию $i$, По возрастанию $j$
* $dp[n][m]$
